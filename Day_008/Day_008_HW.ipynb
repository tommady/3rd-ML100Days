```python
f_income_total_mean = df.query('CODE_GENDER == "F"')["AMT_INCOME_TOTAL"].mean()
f_income_total_std = df.query('CODE_GENDER == "F"')["AMT_INCOME_TOTAL"].std()
f_credit_mean = df.query('CODE_GENDER == "F"')["AMT_CREDIT"].mean()
f_credit_std = df.query('CODE_GENDER == "F"')["AMT_CREDIT"].std()
f_annuity_mean = df.query('CODE_GENDER == "F"')["AMT_ANNUITY"].mean()
f_annuity_std = df.query('CODE_GENDER == "F"')["AMT_ANNUITY"].std()

m_income_total_mean = df.query('CODE_GENDER == "M"')["AMT_INCOME_TOTAL"].mean()
m_income_total_std = df.query('CODE_GENDER == "M"')["AMT_INCOME_TOTAL"].std()
m_credit_mean = df.query('CODE_GENDER == "M"')["AMT_CREDIT"].mean()
m_credit_std = df.query('CODE_GENDER == "M"')["AMT_CREDIT"].std()
m_annuity_mean = df.query('CODE_GENDER == "M"')["AMT_ANNUITY"].mean()
m_annuity_std = df.query('CODE_GENDER == "M"')["AMT_ANNUITY"].std()

pdf = pd.DataFrame([[f_income_total_mean, m_income_total_mean], 
                    [f_credit_mean, m_credit_mean], 
                    [f_annuity_mean, m_annuity_mean], 
                    [f_income_total_std, m_income_total_std], 
                    [f_credit_std, m_credit_std], 
                    [f_annuity_std, m_annuity_std]], index=['AMT_INCOME_TOTAL_MEAN', 'AMT_CREDIT_MEAN', 'AMT_ANNUITY_MEAN', 'AMT_INCOME_TOTAL_STD', 'AMT_CREDIT_STD', 'AMT_ANNUITY_STD'], columns=pd.Index(['Female', 'Male'], name="Genus"))
pdf.plot(kind='bar')
```
